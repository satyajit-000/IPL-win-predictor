# IPL winner Prediction

## Importing Required Libraries

```python
import pandas as pd
import numpy as np
```

## Reading Datasets

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
match = pd.read_csv('IPL_Matches_2008_2022.csv')
```

```python
match.head()
```

```python
delivery.head()
```

```python
match.shape, delivery.shape
```

## Calculating Total Score

```python
total_score_df = delivery.groupby(['ID','innings']).sum()['total_run'].reset_index()
total_score_df.head()
```

## Extraction of Target Score

```python
total_score_df = total_score_df[total_score_df['innings'] == 1]
total_score_df.head()
```

## Get required DataFrame by Merging

```python
match_df = match.merge(total_score_df[['ID','total_run']], on = 'ID')
match_df.head()
```

## Teams inside our Datasets

```python
sorted(match_df['Team1'].unique())
```

## Teams currently playing IPL

```python
teams = [
    'Chennai Super Kings',
    'Delhi Capitals',
    'Gujarat Titans',
    'Kolkata Knight Riders',
    'Lucknow Super Giants',
    'Mumbai Indians',
    'Punjab Kings',
    'Rajasthan Royals',
    'Royal Challengers Bangalore',
    'Sunrisers Hyderabad'
]
```

## Replacing the old Team with there new Team name

```python
match_df['Team1'] = match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
# match_df['Team1'] = match_df['Team1'].str.replace('Gujarat Lions','Gujarat Titans')
match_df['Team1'] = match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team1'] = match_df['Team1'].str.replace('Kings XI Punjab','Punjab Kings')

match_df['Team2'] = match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')
# match_df['Team2'] = match_df['Team2'].str.replace('Gujarat Lions','Gujarat Titans')
match_df['Team2'] = match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2'] = match_df['Team2'].str.replace('Kings XI Punjab','Punjab Kings')
```

```python
delivery['BattingTeam'] = delivery['BattingTeam'].str.replace('Delhi Daredevils','Delhi Capitals')
# delivery['BattingTeam'] = delivery['BattingTeam'].str.replace('Gujarat Lions','Gujarat Titans')
delivery['BattingTeam'] = delivery['BattingTeam'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
delivery['BattingTeam'] = delivery['BattingTeam'].str.replace('Kings XI Punjab','Punjab Kings')
```

## Selecting the the only teams those are currently playing in IPL

```python
match_df = match_df[match_df['Team1'].isin(teams)]
match_df = match_df[match_df['Team2'].isin(teams)]
match_df.head()
```

```python
delivery = delivery[delivery['BattingTeam'].isin(teams)]
delivery.head()
```

## Getting delivery dataframe by merging delivery match_df and delivery

```python
delivery_df = match_df[['ID','City','WinningTeam','total_run','Team1','Team2']].merge(delivery, on = 'ID')
```

## Only selecting the 2nd innings for prediction

```python
delivery_df=delivery_df[delivery_df['innings'] == 2]
```

```python
delivery_df.head()
```

## Getting the Bowling Team

```python
delivery_df[['Team1','Team2','BattingTeam']].sample(10)
```

```python
delivery_df['BowlingTeam'] = np.where(
    delivery_df['BattingTeam'] == delivery_df['Team2'], 
    delivery_df['Team1'], delivery_df['Team2'])
```

```python
delivery_df.tail()
```

## Calculating *current_score*, *runs_left*, *balls_left*, *wicket_left*, *current_runrate* & *required_runrate*

```python
delivery_df['current_score'] = delivery_df.groupby('ID')['total_run_y'].cumsum().astype('int')
```

```python
delivery_df['runs_left'] = delivery_df['total_run_x'] - delivery_df['current_score'] + 1
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['overs']*6 + delivery_df['ballnumber'])
```

```python
delivery_df['player_out'] = delivery_df['player_out'].fillna('0')
delivery_df['player_out'] = delivery_df['player_out'].apply(lambda x: x if x == '0' else '1')
delivery_df['player_out'] = delivery_df['player_out'].astype('int')
wickets = delivery_df.groupby('ID')['player_out'].cumsum().values
delivery_df['wicket_left'] = 10 - wickets
```

```python
delivery_df['crr'] = delivery_df['current_score']/((120 - delivery_df['balls_left']) / 6)
```

```python
delivery_df['rrr'] = delivery_df['runs_left'] / ((delivery_df['balls_left']) / 6)
```

```python
delivery_df.head()
```

## Calculating result of the match *(win/loss)*

```python
delivery_df['result'] = delivery_df.apply(lambda row: 1 if row['BattingTeam'] == row['WinningTeam'] else 0,axis = 1)
```

## Extracting the required columns to train our model

```python
final_df = delivery_df[['City', 'BattingTeam', 'BowlingTeam', 'runs_left', 'balls_left','wicket_left','total_run_x','crr','rrr','result']]
final_df
```

## Handelling the inifinity and Nan values

```python
final_df = final_df[final_df['balls_left'] != 0]
final_df.dropna(inplace = True)
final_df
```

## Saving the final_df as csv file for UI/server design using streamlit

```python
final_df.to_csv('final_dataset.csv', index = False)
```

## Suffle the datasets to prevent from bias

```python
final_df = final_df.sample(final_df.shape[0])
final_df.head()
```

## Extracting X and y to train our model

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
```

## Performing train_test_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 191)
```

```python
X_train
```

## Implementation of ColumnTransformer, OneHotEncoder and Pipeline using *LogisticRegression*

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

encode = ColumnTransformer([
    ('encode', OneHotEncoder(sparse=False, drop='first'), ['City', 'BattingTeam', 'BowlingTeam'])
], remainder='passthrough')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
pipe = Pipeline(steps = [
    ('encode',encode),
    ('model',LogisticRegression(solver = 'liblinear'))
])
```

## Training the Model and predicting output

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

## Accuracy Score

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

## Predict Probability

```python
pipe.predict_proba(X_test)[np.random.randint(0, X_test.shape[0])]
```

## Saving the model using pickle

```python
import pickle
pickle.dump(pipe,open('IPL_model.pkl', 'wb'))
```
